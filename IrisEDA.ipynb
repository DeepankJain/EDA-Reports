Step 2: Objective: Given the sepal length, petal length, sepal width, and petal width, predict the species of the flower.

```
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
#loading the dataset
iris = sns.load_dataset('iris')
iris.head()
```

```
#number of rows and columns in the dataset
iris.shape
```

```
#name of columns
iris.columns
```

```
#type of attributes in the dataset
iris.info()
```

```
numerical = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
categorical = ['species']
```

```
#checking for missing values
iris.isnull().sum()
```

```
#Univariate Analysis
#Non graphical
iris.describe()
```

```
iris.describe(include = ['object'])
```

```
#count the datapoints for each class
iris['species'].value_counts()
```

```
from collections import Counter
Counter(iris['species'])
```

```
#graphical
#histogram
plt.subplot(221)
sns.distplot(iris['sepal_length'], color = 'b')
plt.subplot(222)
sns.distplot(iris['sepal_width'], color = 'g')
plt.subplot(223)
sns.distplot(iris['petal_length'], color = 'y')
plt.subplot(224)
sns.distplot(iris['petal_width'], color = 'k')
plt.show()
```

```
#creating box plot
sns.boxplot(iris['sepal_length'], color = 'g')
plt.show()
```

```
#box plot grouped by a categorical variable
sns.boxplot(data = iris, x = 'species', y = 'sepal_length')
plt.show()
```

```
#Bar plot of categorical attribute
sns.countplot(iris['species'])
plt.show()
```

```
#Bivariate Analysis
#Contingency Table
pd.crosstab(iris['sepal_length'], iris['species'])
```

```
#Correlation matrix
corr_matrix = iris[numerical].corr()
print(corr_matrix)
```

```
sns.heatmap(corr_matrix)
```

```
#Scatter Plot
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris, kind = 'scatter')
plt.show()
```

```
#KDE Plot
sns.jointplot(data = iris, x = 'sepal_width', y = 'petal_width', kind = 'kde')
plt.show()
```

```
#Pair Plot
sns.pairplot(iris[numerical])
plt.show()
```

```
#LM Plot
sns.lmplot(x = 'sepal_length', y = 'petal_length', data = iris, hue = 'species', fit_reg = True)
plt.show()
```
